## **Exploración inicial**

```python
#Importo librerías
#!pip install pandas 
#!pip install seaborn 
#!pip install matplotlib 
#!pip install numpy
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Carga del dataset y se realiza una inspección inicial
df_pinguinos = sns.load_dataset("penguins")
```

```python
#Inspección inicial
print("Primeras filas del dataset:")
print(df_pinguinos.head())

print("Información del dataset:")
print(df_pinguinos.info())

print("Descripción estadística del dataset:")
print(df_pinguinos.describe())
```

```python
#Manejo de valores faltantes
print("Valores faltantes en cada columna:")
print(df_pinguinos.isnull().sum())

df_pinguinos_limpio = df_pinguinos.dropna()
print("Dataset después de eliminar filas con valores faltantes:")
print(df_pinguinos_limpio.isnull().sum())
```

## **Visualizaciones requeridas**

### Distribución de una variable numérica (Matplotlib)

```python
#Histograma utilizando matplotlib para visualizar la distribución de la variable flipper_length_mm
plt.figure(figsize=(8, 6))
plt.hist(df_pinguinos_limpio['flipper_length_mm'], bins=25, color='cornflowerblue', edgecolor='black')
plt.title('Distribución de la Longitud de aleta (flipper_length_mm)')
plt.xlabel('Longitud de aleta (mm)')
plt.ylabel('Frecuencia')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

En este gráfico de distribución de la longitud de la aleta se puede observar que la mayor cantidad de pinguinos poseen una longitud de aleta que se acerca a los 190mm. De ahí se visualiza que también existe una gran cantidad de pinguinos que tienen una longitud de aleta cercana a los 210 mm.

### Frecuencia de una variable categórica

```python
#Gráfico de barras utilizando seaborn para mostrar el número de pinguinos por especie
plt.figure(figsize=(8, 6))
sns.countplot(data=df_pinguinos_limpio, x='species', palette='winter')
plt.title('Número de Pingüinos por Especie')
plt.xlabel('Especie')
plt.ylabel('Número de Pingüinos')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

En este gráfico de barras se puede identificar que las especies con mayor número de pinguinso es la especie Adelie, seguida de la especie Gentoo y finalmente la especie Chinstrap.

### Relación entre dos variables numéricas

```python
#Gráfico de disperción con matplotlib para explorar la relación entre bill_length_mm y bill_depth_mm
plt.figure(figsize=(8, 6))
plt.scatter(df_pinguinos_limpio['bill_length_mm'], df_pinguinos_limpio['bill_depth_mm'], color='cornflowerblue', alpha=0.7)
plt.title('Relación entre Longitud del Pico y Profundidad del Pico')
plt.xlabel('Longitud del Pico (mm)')
plt.ylabel('Profundidad del Pico (mm)')
plt.grid(alpha=0.75)
plt.show()
```

En este gráfico se pueden observar varios grupos de puntos, lo que sugiere que diferentes especies de pingüinos tienen combinaciones distintas de longitud y profundidad de pico. Por ejemplo, hay un grupo con picos más largos y menos profundos, y otro con picos más cortos y más profundos.

### Comparación de una variable numérica entre categorías

```python
#Diagrama de caja utilizando seaborn para comparar la distribución de masa corporal entre las especies
plt.figure(figsize=(8, 6))
sns.boxplot(data=df_pinguinos_limpio, x='species', y='body_mass_g', palette='winter')
plt.title('Distribución de Masa Corporal por Especie')
plt.xlabel('Especie')
plt.ylabel('Masa Corporal (g)')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

Este diagrama de caja compara la distribución de la masa corporal entre las diferentes especies de pingüinos. Se observa claramente que la especie Gentoo tiene una masa corporal significativamente mayor que las otras especies. La especie Adelie muestra una masa corporal menor que las otras pero con un rango intercuartílico más amplio que Chinstrap.

```python
#Gráfico de violín con seaborn para la misma comparación y diferenciación por sex con el parámetro hue
plt.figure(figsize=(8, 6))
sns.violinplot(data=df_pinguinos_limpio, x='species', y='body_mass_g', hue='sex', palette='winter')
plt.title('Distribución de Masa Corporal por Especie y Sexo')
plt.xlabel('Especie')
plt.ylabel('Masa Corporal (g)')
plt.grid(axis='y', alpha=0.75)
plt.legend(title='Sexo')
plt.show()
```

Este gráfico de violín muestra la distribución de la masa corporal por especie diferenciando por sexo. Se puede apreciar que los machos de cada especie tienden a tener una masa corporal ligeramente superior que las hembras. Para la especie Gentoo, la diferencia de masa corporal entre machos y hembras es más pronunciada que en las otras especies.

### Mapa de calor de correlaciones

```python
# Matriz de correlación de variables numéricas
var_numericas = df_pinguinos_limpio.select_dtypes(include=['number'])
matriz_correlacion = var_numericas.corr()
print("Matriz de correlación:")
print(matriz_correlacion)
```

```python
#Mapa de calor con seaborn para visualizar la matriz de correlación
plt.figure(figsize=(6, 8))
sns.heatmap(matriz_correlacion, annot=True, fmt=".2f", cmap='winter', square=True, cbar_kws={"shrink": .8})
plt.title('Mapa de Calor de la Matriz de Correlación')
plt.show()
```

Este mapa de calor muestra la matriz de correlación entre las variables numéricas del conjunto de datos de pingüinos. Los valores cercanos a 1 o -1 indican una correlación fuerte, mientras que los valores cercanos a 0 indican una correlación débil.

### Figura con múltiples gráficos

```python
#Crear las subparcelas
fig, axes = plt.subplots(1, 2, figsize=(10, 8))

#Subparcela 1 que muestra la longitud de pico de los primeros 20 pinguinos
primeros_20 = df_pinguinos_limpio.head(20)
axes[0].plot(np.arange(len(primeros_20)), primeros_20['bill_length_mm'], marker='o', linestyle='-', color='cornflowerblue')
axes[0].set_title('Longitud de Pico de los Primeros 20 Pingüinos')
axes[0].set_xlabel('Índice')
axes[0].set_ylabel('Longitud del Pico (mm)')
axes[0].grid(True)

#Subparcela 2 gráfico de dispersión que muestra flipper_length_mm vs body_mass_g, coloreando los puntos por especia
sns.scatterplot(
    data=df_pinguinos_limpio,
    x='flipper_length_mm',
    y='body_mass_g',
    hue='species',
    ax = axes[1],
    palette='winter',
)
axes[1].set_title('Relación entre Longitud de Aleta y Masa Corporal por Especie')
axes[1].set_xlabel('Longitud de Aleta (mm)')
axes[1].set_ylabel('Masa Corporal (g)')
axes[1].legend(title='Especie')
axes[1].grid(True)

plt.tight_layout()
```

La subparcela 1 muestra la longitud del pico de los primeros 20 pingüinos y permite observar la variabilidad de la longitud del pico en una pequeña muestra secuencial y no se puede identificar una tendencia marcada en este conjunto de datos.
La subparcela 2 muestra al relación entre la longitud de la aleta y la masa corporal, coloreando los puntos según la especie. Este gráfico muestra una fuerte correlación con referencia a que los pingüinos con aletas más largas tienden a ser más pesados.
